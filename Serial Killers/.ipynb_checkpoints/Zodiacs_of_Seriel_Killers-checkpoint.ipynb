```python
wiki_wiki = wikipediaapi.Wikipedia('Serial_Killer_Zodiac_Signs (cferandel3@gmail.com)','en')

page_py = wiki_wiki.page('List of serial killers in the United States')
```

```python
months = {'january':1, 'february':2, 'march':3,'april':4,'may':5, 
          'june':6,'july':7, 'august':8,'september':9,'october':10,
          'november':11,'december':12}
```

```python
def find_date(aString):
    birth_date_position = 1000
    birth_month = ""
    birth_day = 0
    birth_year = 0
    comma_position = aString.find(",")
    if comma_position == -1:
        return
    for month in months.keys():
        position = aString.lower().find(month)
        if position >= 0 and position < birth_date_position:
            birth_date_position = position
            birth_month = month
    birth_day = aString[birth_date_position + len(birth_month) + 1:comma_position]
    birth_year = aString[comma_position + 2: comma_position + 6]
    try:
        return(date(int(birth_year), int(months[birth_month]), int(birth_day)))
    except:
        return
    
    
```

```python
with_birthday["Birthday"] = with_birthday["Birthday"].apply(find_date)
```

```python
# Define the start dates for each zodiac sign
    
sign_start_dates = {
        "Capricorn": datetime.strptime('12-22', '%m-%d'),
        "Sagittarius": datetime.strptime('11-22', '%m-%d'),
        "Scorpio": datetime.strptime('10-23', '%m-%d'),
        "Libra": datetime.strptime('09-23', '%m-%d'),
        "Virgo": datetime.strptime('08-23', '%m-%d'),
        "Leo": datetime.strptime('07-23', '%m-%d'),
        "Cancer": datetime.strptime('06-21', '%m-%d'),
        "Gemini": datetime.strptime('05-21', '%m-%d'),
        "Taurus": datetime.strptime('04-20', '%m-%d'),
        "Aries": datetime.strptime('03-21', '%m-%d'),
        "Pisces": datetime.strptime('02-19', '%m-%d'),
        "Aquarius": datetime.strptime('01-20', '%m-%d')
    }
```

```python
def get_zodiac_sign(date):

    # Determine the zodiac sign
    for sign, start_date in sign_start_dates.items():
        if date.month > start_date.month:
            return sign
        elif (date.month == start_date.month) & (date.day >= start_date.day):
            return sign

    return("Capricorn")
```

```python
with_birthday['Zodiac'] = with_birthday['Birthday'].apply(get_zodiac_sign)
```
